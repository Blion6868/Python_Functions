```python
#Monte Carlo Simulation for ad revenue

import random
import matplotlib.pyplot as plt

def monte_carlo(num_sims): #determine number of simulations
    total_cost_samples = []
    
    for _ in range(num_sims):
        print_ad_cost = random.uniform(50000, 70000) #random print ad costs; can be adjusted based on domain knowledge
        digital_ad_cost = random.uniform(100000, 200000) #random digital ad costs; can be adjusted based on domain knowledge
        ad_duration = random.uniform( 12, 18) #ad duration
        
        total_cost = print_ad_cost + digital_ad_cost
        total_cost_samples.append(total_cost)
        
    return total_cost_samples

       

if __name__ == "__main__":
    num_sims = int(input("Enter the number of MC simulations: "))
    cost_samples = monte_carlo(num_sims)
    
    mean_cost = sum(cost_samples) / len(cost_samples)
    std_dev = (sum((x - mean_cost)**2 for x in cost_samples) / (len(cost_samples) - 1))**0.5
    
    print(f"Estimated Mean Total Ad Project Cost: ${mean_cost:.2f}")
    print(f"Estimated Standard Deviation: ${std_dev:.2f}")
    
     # Plotting the data with customizations
    plt.figure(figsize=(10, 6))  # Adjust the figure size
    
    # Plot the data as a blue line with circular markers
    plt.plot(cost_samples, color='b', marker='o', linestyle='-')
    
    # Add labels and title
    plt.xlabel("Simulation")
    plt.ylabel("Total Ad Cost")
    plt.title("MC Simulation Ad Costs")
    
    # Add a grid for better readability
    plt.grid(True)
    
    # Display the mean cost as a horizontal line
    plt.axhline(y=mean_cost, color='r', linestyle='--', label=f"Mean Cost: ${mean_cost:.2f}")
    
    # Display the legend
    plt.legend()
    
    # Show the plot
    plt.show()
```

```python
#the larger the simulation samples, the closer the result will match the true value. We also limit the standard deviation, which
#will result in better results. 
```
